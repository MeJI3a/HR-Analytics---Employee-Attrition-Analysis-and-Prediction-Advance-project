```python
duplicates = df.duplicated()
print("Duplicate rows:\n", duplicates)
```

```python
duplicate_rows = df[df.duplicated()]
print("Duplicate rows:\n", duplicate_rows)
```

```python
df['YearsWithCurrManager'].fillna(df['YearsWithCurrManager'].median(), inplace=True)
```

```python
# Fill missing values in 'YearsWithCurrManager' column by assigning it back to the column directly
df['YearsWithCurrManager'] = df['YearsWithCurrManager'].fillna(df['YearsWithCurrManager'].median())
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['BusinessTravel'].replace({'TravelRarely': 'Travel_Rarely'}, inplace=True)
df.drop(columns=['Over18', 'StandardHours', 'EmployeeCount'], inplace=True)
```

```python
# Replace 'TravelRarely' with 'Travel_Rarely' in 'BusinessTravel' column
df['BusinessTravel'] = df['BusinessTravel'].replace({'TravelRarely': 'Travel_Rarely'})
```

```python
categorical_cols = ['AgeGroup', 'Attrition', 'BusinessTravel', 'Department', 'EducationField', 
                    'Gender', 'JobRole', 'MaritalStatus', 'SalarySlab']
df[categorical_cols] = df[categorical_cols].astype('category')
```

```python
rating_columns = ['EnvironmentSatisfaction', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 
                  'RelationshipSatisfaction', 'WorkLifeBalance', 'PerformanceRating']
for col in rating_columns:
    print(f"{col} range: {df[col].min()} - {df[col].max()}")
```
